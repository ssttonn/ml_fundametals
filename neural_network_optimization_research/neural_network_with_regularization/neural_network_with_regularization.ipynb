```python
parameters = he_initialize_parameters(layer_dims)

parameters, costs = nn_model(train_X, train_Y, layer_dims, parameters, activations, number_of_iterations=100, learning_rate=0.1, print_cost=True)
parameters
```

```python
predictions_train = predict(parameters, activations, train_X)

metrics(train_Y[0], predictions_train[0])
```

```python
predictions_test = predict(parameters, activations, test_X)
metrics(test_Y[0], predictions_test[0])
```
